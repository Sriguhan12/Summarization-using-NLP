# Storing the content in doc variable

```python
doc = nlp(text)
```

# Tokenization
    Each Doc consists of individual tokens, and we can iterate over them. Now we will make a list of tokens called tokens.

```python
tokens = [token.text for token in doc]
print(tokens)
```

# Punctuations

       We can see that all the punctuation marks and special characters are included in the tokens. Now we will remove them. punctuation contains a string of all the punctuations but it does now conatin \n. So we will add \n in punctuation.

```python
punctuation = punctuation + '\n'
punctuation
```

# Word frequency table

      Now we will make the word frequency table. It will contain the number of occurrences of all the distinct words in the text which are not punctuations or stop words. We will create a dictionary named word_frequencies.

```python
word_frequencies = {}
for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
                
print(word_frequencies)
```

# Max frequency

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

# Normalization of frequencies
    Divide each frequency value in word_frequencies with the max_frequency to normalize the frequencies.

```python
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency

print(word_frequencies)
```

# Sentence Tokenization

```python
sentence_tokens = [sent for sent in doc.sents]
print(sentence_tokens)
```

# Sentence scores
    The sentence score for a particular sentence is the sum of the normalized frequencies of the words in that sentence. All the sentences will be stored with their score in the dictionary sentence_scores.

```python
sentence_scores = {}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]
                
sentence_scores
```

# Sentence reduction
          Since the requirement is only 500 words we are going to reduce the content to 7% of the original content by using nlargest from heapq package

```python
from heapq import nlargest
```

```python
select_length = int(len(sentence_tokens)*0.070)
select_length
```

# Final Summarization

```python
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
summary
```
